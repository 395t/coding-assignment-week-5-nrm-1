```python
%matplotlib inline
import json
import glob
import matplotlib
from matplotlib import pyplot as plt

plt.style.use("ggplot")
```

```python
curve_files = glob.glob("src/stats/STL10*train_metrics.json")[:-1]
del curve_files[-3:-1]
del curve_files[0:3]
curve_files
```

```python
for filename in curve_files:
    with open(filename, "rb") as f:
        cont = json.load(f)
    curve = [dp["loss"] for dp in cont.values()]
    plt.plot(curve, label=filename.split("/")[-1][:-19])
plt.legend()
```
